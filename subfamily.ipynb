```python
import requests
import re
from threading import Thread
import queue
from threading import Semaphore
from lxml import etree
import json
import ssl
```

```python
prefix = "http://www.cazy.org/"
fivefamilies = ["Glycoside-Hydrolases.html"]
#fivefamilies = ["Auxiliary-Activities.html"]

rx_subfamily = re.compile(r'<a id="separ_subfam" href="(\w+).html">')

for family in fivefamilies:
    address = prefix + family
    
    page = requests.get(address, verify=False).content.decode('iso8859-1')

    #print (page)
    subfamilies = rx_subfamily.findall(page)

    for subfamily in subfamilies:
        print (f"{prefix}{subfamily}.html")
```
